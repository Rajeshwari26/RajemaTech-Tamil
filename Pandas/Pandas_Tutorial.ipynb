## Why Pandas - To analyze, cleaning, exploring, and manipulating data
**Data Structures:** Series and DataFrame


**Series**=> A one-dimensional array-like object

**DataFrame**=>A two-dimensional table with rows and columns


```python
# Series
import pandas as pd
import numpy as np

data = pd.Series([1, 3, 5, np.nan, 6, 8])
print(data)
```

```python
# Dataframe

data = {'Name': ['Raj', 'Rajesh', 'Ram','Ravi','vidya','Sona','Ragavi','Saran'],
        'Age': [25, 30, 35,20,25,22,26,np.nan],
        'City': ['chennai', 'Bangalore', 'kerala','chennai','kashmir','delhi','mysore','kerala']}

df = pd.DataFrame(data)

```
